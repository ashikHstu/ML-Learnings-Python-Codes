# Seaborn :

Data visualization library

Importing the libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

<b>Note : Seaborn has some built-in datasets </b>

```python
#total bull vs tip dataset
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# # setting a theme for the plots
# sns.set_theme()
```

```python
# Visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#load the iris dataset
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

# Scatter Plot

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
# Loading the titanic dataset
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.shape
```

## Count Plot

```python
sns.countplot(x='class',data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

# Bar Chart

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
# House price dataset
from sklearn.datasets import load_boston
house_boston=load_boston()
house=pd.DataFrame(house_boston.data,columns=house_boston.feature_names)
house['Price']=house_boston.target
```

```python
house.head()
```

```python
print(house_boston)
```

# Distribution Plot

```python
sns.distplot(house['Price'])
```

# Corelation : 
<br> 
- Positive Correlation <br>
- Negative Correlation

<b> Head Map </b>

```python
correlation=house.corr()
```

```python

# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
